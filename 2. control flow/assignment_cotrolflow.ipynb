```python
#find no.is positive or not 

Num = int(input("Enter a Number :"))

if Num>0:
    print("Positive No.")
elif Num<0:
    print("Negative No.")
else:
    print("Number is Zero")
```

```python
# No is positive(even or odd) or Negative
 
Num = int(input("Enter a Number :"))

if Num>0:
    if Num%2==0:
       print("Positive Even No.")
    else:
        print("Positive Odd No.")
elif Num<0:
    if Num%-2==0:
       print("Negative Even No.")
    else:
        print("Negative Odd No.")
else:
    print("Number is Zero")
```

```python
#Write a program that prints all the numbers from 1 to 10 using a while loop.

Num = int(input("Enter a Number :"))
count=0
while count<=Num:
    print(count)
    count+=1
```

```python
#Write a program that asks the user to input numbers until they input 0. The program should print the sum of all the input numbers.

total = 0

while True:
    Numbers = float(input("Enter the Numbers to add the all up (0 to stop): "))
    if Numbers == 0:
        break
    total += Numbers
print(f"The sum of all the numbers is {total}")
```

```python
#Write a program that prints all the numbers from 1 to 10 except 5 using a for loop and continue statement.

for i in range(1,11):
    if i==5:
        continue
    print(i)
```

```python
#Write a program that calculates the factorial of a number input by the user using a while loop.

num = float(input("Enter the no. : "))
fact = 1
i=1
while i<=num:
    fact = fact*i
    i +=1
print(fact) 
```
