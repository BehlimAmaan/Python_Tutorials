```python
# Modeling  a bank Account 

class BankAccount:
    
    def __init__(self, Owner, Balance=0):
        self.Ownwe=Owner
        self.Balance=Balance
    def deposite(self, Amount):
        self.Balance+=Amount
        print(f"{Amount} deposited and the balance is {self.Balance}")
    def withdraw(self,Amount):
        if Amount>self.Balance:
            print("Insufficient Balance")
        else:
            self.Balance-=Amount
            print(f"Total withdraw is {Amount} balance remaining is {self.Balance}")
    def get_balance(self):
        return self.Balance

Account = BankAccount("Amaan", 100000)
print(Account)
print(Account.Balance)
print(Account.withdraw(10000))
print(Account.deposite(2000))

```
