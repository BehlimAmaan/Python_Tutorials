

Project: Student Record Management System (Python, Data Structures, Functions)

add_student() – to add one student

search_student() – find student using roll no

update_marks() – change marks

delete_student() – remove student

display_all() – show all students

menu() – show options


```python
students = [
    {'name': 'Amaan', 'Roll': 5, 'Grade':'A'},
    {'name': 'Zaid', 'Roll': 3, 'Grade':'b'},
    {'name': 'Aftab', 'Roll': 21, 'Grade':'A'}
]
```

```python
def add_student():
    
    name = str(input("Enter Student Name :"))
    Roll  = int(input("Enter Students Roll Number :"))
    Grade = str(input("Enter Grades of Student"))
    
    student = {'name':name,'Roll':Roll,'Grade':Grade}
    students.append(student)
    print("Student Added Successfully")
```

```python
def search_student():
    Roll = int(input("Enter Student Roll: "))
    for s in students:
        if s['Roll']== Roll:
            print(f"Student found: {s}")
            return 
        else : 
            print(f"Student with Roll No. {s} Not Found ")
```

```python
def update_Grade():
    Roll = int(input("Enter Student Roll: "))
    for s in students:
        if s['Roll'] == Roll:
            new_Grade = int(input("Enter Studens New Grade: "))
            s['Grade'] = new_Grade
            return
        else:
            print(f"Student with Roll No. {s} Not Found ")
```

```python
def delete_student():
    Roll = int(input("Enter Student Roll: "))
    for s in students:
        if s['Roll'] == Roll:
            students.remove(s)
            return
        else:
            print("Student Not Found ")
```

```python
def display_all():
    if not students:
        print("No Data Found ")
        return
    for s in students:
         print(s)
        
```

```python
def menu():
    while True:
        print(".........Welcom to The student Management.......")
        print("1. add a Student ...")
        print("2. Search for the Students")
        print("3. Update the Marks of the Students....")
        print("4. Delete a Student from the Table....")
        print("5. Display all the students.....")
        print("6. Exit....!!!")
    
        choice = int(input("Enter Your choice :"))
        
        if choice == 1:
            return add_student()
        elif choice == 2:
            return search_student()
        elif choice == 3:
            return update_Grade()
        elif choice == 4:
            return delete_student()
        elif choice == 5:
            return display_all()
        elif choice == 6:
            print("Exiting...")
            break
        else: 
            print("No Data Found!!!!")
menu()
            
            
```
