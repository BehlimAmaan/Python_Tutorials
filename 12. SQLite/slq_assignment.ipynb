2. Write a Python function to create a table named `employees` with columns `id` (integer), `name` (text), `age` (integer), and `department` (text) in the `test.db` database.



```python
def create_table():
    conn =sqlite3.connect('Amaan.db')
    cursor = conn.cursor()
    cursor.execute('''
            CREATE TABLE IF NOT EXISTS Employee(
            id INTEGER PRIMARY KEY,
            name TEXT NOT NULL,
            age INTEGER,
            department TEXT
            )
    ''')
    conn.commit()
    conn.close()
    print("Employee Table Craeted Successfully")
create_table()
```

### Assignment 2: Inserting Data

1. Write a Python function to insert a new employee into the `employees` table.

```python
def add_employee(id,name,age,departement):
    conn =sqlite3.connect('Amaan.db')
    cursor=conn.cursor()
    cursor.execute('''
        INSERT INTO Employee(id,name,age,department)
        VALUES(?,?,?,?)
    ''',(id,name,age,departement))
    conn.commit()
    conn.close()
    print("Employee Added successfully")
add_employee(1,'Zaid',22,'Data Analyst')
```

### Assignment 3: Querying Data

1. Write a Python function to fetch and display all records from the `employees` table.

```python
def print_all():
    conn = sqlite3.connect('Amaan.db')
    cursor=conn.cursor()
    cursor.execute('SELECT * FROM Employee')
    records = cursor.fetchall()
    conn.close()
    for record in records:
        print(record)
print_all()
```

2. Write a Python function to fetch and display all employees from a specific department.

```python
def select_from_department(department):
    conn = sqlite3.connect('Amaan.db')
    cursor=conn.cursor()
    cursor.execute('SELECT * FROM Employee WHERE department=?',(department,))
    records = cursor.fetchall()
    conn.close()
    for record in records:
        print(record)
        print("Fetch Successfully")
select_from_department('Amul')
```

### Assignment 4: Updating Data

1. Write a Python function to update the department of an employee based on their `id`.

```python
def update_department(id,new_department):
    conn = sqlite3.connect('Amaan.db')
    cursor = conn.cursor()
    cursor.execute('''
        UPDATE Employee
        SET department = ?
        WHERE id = ?
    ''',(new_department,id))
    conn.commit()
    conn.close()
update_department(3,'Bussiness')
```
