#### Example 1: Temperature Conversion

```python
def temp_coversion(tem,unit):
    """To convert from celcious to ferhanhite"""
    if unit == 'C':
        return (tem * 9/5) + 32
    elif unit == 'F':
        return (tem - 32) * 9/5
    elif unit == 'K':
        return tem + 273.15
    else :
        return None
    
temp= int(input("Enter the temperature and unit"))
units = str(input("Enter the Unit ('C','F','K') :"))
print(temp_coversion(temp ,units))    
    
```

##### Example 2: Password Strength Checker

```python
def is_password_strong(password):
    if len(password)<8:
        return False
    elif not any(char.isdigit() for char in password):
        return False
    elif not any(char.islower() for char in password):
        return False
    elif not any(char.isupper() for char in password):
        return False
    elif not any(char in '#@!*&|?/' for char in password):
        return False
    else:
        return True
    
pas = input("Enter to check passwor strenght :")
print(is_password_strong(pas))
    
```

##### Example 3: Calculate the Total Cost Of Items In a Shopping Cart

```python
def total_items_cost(cart):
    total_cost = 0
    for items in cart:
        total_cost +=items["price"]*items["qty"]
    return total_cost

cart = [
    {"item_name":"laptop", "price": 50000, "qty":4},
    {"item_name":"Mobile", "price": 27000, "qty":2},
    {"item_name":"RAM", "price": 50000, "qty":6},
    {"item_name":"GPU", "price": 300000, "qty":1}
]

print(total_items_cost(cart))
```

##### Example 4: Check IF a String Is Palindrome

```python
def palindrome(string):
    if string==string[::-1]:
        return True
    else:
        return False
    
str_n = (input("Enter Your String  :"))
print(f"The palindrome of String is : {palindrome(str_n)}")
        
```

##### Example 5: Calculate the factorials of a number using recursion

```python
def factorial(n):
    if n==0:
        return 1
    else:
       return n * factorial(n-1)
   
Num = int(input("Enter Your No. : "))
print(f"Factorial of a  number is : {factorial(Num)}")
```
